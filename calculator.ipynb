```python
def summ (a , b):
    return a+b
```

```python
def sub (a , b):
     return a-b
```

```python
def multi (a , b):
     return a*b
```

```python
def div (a , b):
     return a/b
```

```python
def floordiv (a , b):
     return a//b
```

```python
def modulo (a , b):
     return a%b
```

```python
def power (a , b):
     return a**b
```

```python
def absolute (a):
    return (a**2)**0.5
```

```python
def nthroot (num, root):
    n=num**(1/root)
    return n
```

```python
def factorial (n):
    return 1 if (n==1 or n==0) else n*factorial (n-1)
```

```python
from math import sqrt
#Quadratic equation   y= ax^2+bx+c 

def Quadreq (a,b,c):
#1-enter coefficients 
    a,b,c = eval (input ("enter coefficients"))
#2- calcul discriminant 
    discr = b*b - 4*a*c
    if discr > 0 :
        sol1= (-b+sqrt(discr))/(2*a)
        sol2= (-b-sqrt(discr))/(2*a)
        print (f'solution: {sol1},{sol2}')
    else : 
        print ('no solution')
```
